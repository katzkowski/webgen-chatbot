# Image Grid

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import random
import numpy as np

from pathlib import Path
from dotenv import load_dotenv
from keras.preprocessing.image import load_img

load_dotenv()
DATA_PATH = Path(os.getenv("DATA_PATH"))
```

## Load images

```python
# path to dataset
# dataset_path = DATA_PATH / "datasets" / "v02_k55_c1_root" / "v02_k55_c1"

target_run = "transfer_m70_v02_k55_c40-generated-71"
dataset_path = DATA_PATH / "generated" / target_run

# choose n random images
n = 24

# 4, 0
seed = 10
random.seed(a=seed)
images = random.sample([
    image
    for image in os.listdir(dataset_path)
    if os.path.isfile(os.path.join(dataset_path, image))
], k=n)
```

## Create grid as cluster preview

```python
data = [np.array(load_img(dataset_path / img)) for img in images]

size_x, size_y = (8,3)
linewidth = 6


# plot grid
fig, axs = plt.subplots(size_y,size_x, figsize=(80, 30))

for i in range(0,size_y):
    for j in range(0,size_x):
        axs[i,j].imshow(data.pop())
        # axs[i,j].axis("off")

        # add borders
        axs[i,j].spines["top"].set_visible(True)
        axs[i,j].spines["top"].set_linewidth(linewidth)

        axs[i,j].spines["bottom"].set_visible(True)
        axs[i,j].spines["bottom"].set_linewidth(linewidth)

        axs[i,j].spines["right"].set_visible(True)
        axs[i,j].spines["right"].set_linewidth(linewidth)

        axs[i,j].spines["left"].set_visible(True)
        axs[i,j].spines["left"].set_linewidth(linewidth)

        # hide ticks
        axs[i,j].set_xticklabels([])
        axs[i,j].set_yticklabels([])


fig.tight_layout()
plt.subplots_adjust(hspace=0.001, wspace=0.001, bottom=0.001)

plt.show()
```

## Save file

```python
samples_path = DATA_PATH / "samples"
filename = target_run + "_s" + str(seed) + ".jpg"


fig.savefig(samples_path / filename)
```
